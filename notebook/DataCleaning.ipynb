As per above analysis, Rows with missing customer id are removed as they will not be usable for analysis but rows having missing decription are kept and reviewed later. 

**Handling Missing value**

```python
df_copy=df_copy.dropna(subset=['CustomerID']) #removing missing customer ID rows
```

```python
df_copy.shape
```

**Checking Duplicate values**

```python
df_copy[df_copy.duplicated()]
#As here invoiceNo and customer ideas are repeated doesn't means its duplicates.
#We will say duplicates only when completely identical rows acrosss all columns.
```

**Removing Duplicates**

```python
df_copy= df_copy.drop_duplicates()
#we will remove duplicates 
```

**Handling cancelled orders**

```python
df_copy=df_copy[~df_copy['InvoiceNo'].astype(str).str.startswith('C')]

#filtering rows which contains cancellation order 
```

```python
df_copy[(df_copy['Quantity']<=0)]
```

```python
df_copy['InvoiceDate']= pd.to_datetime(df_copy['InvoiceDate'])
```

```python
df_copy['CustomerID']=df_copy['CustomerID'].astype(int)
```
