## Customer Analysis

**Revenue per Customer**

```python
rev_customer = df.groupby('CustomerID')['Total_amt'].sum()
rev_customer
```

**No. of orders per customer**

```python
df.groupby('CustomerID')['InvoiceNo'].nunique()
```

**Avg order value per customer**

```python
order_value = df.groupby('InvoiceNo')['Total_amt'].sum()
avg_order = order_value.groupby(df.set_index('InvoiceNo')['CustomerID']).mean()
```

```python
#sort by revenue
sort= df["Total_amt"].sort_values()
sort
```

**Top 10 customers**

```python
top= df.groupby('CustomerID')['Total_amt'].head(10)
top
```

## Time Analysis

```python

df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceDate'].dt.to_period('M')
```

**Monthly revenue**

```python
amt_month= df.groupby(df['InvoiceDate'].dt.month)['Total_amt'].sum()
amt_month
```

**Monthly Orders**

```python
ord_month= df.groupby(df['InvoiceDate'].dt.month)['InvoiceNo'].sum()
ord_month
```
